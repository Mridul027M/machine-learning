```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
data=pd.read_csv("/content/student_scores.csv")
#print(data)
fea=data.shape[0]
patt=data.shape[1]
print("number of features-",fea)
print("number of patterns-",patt)
range=[]
max=data["Scores"].max()
min=data["Scores"].min()
range.append(min)
range.append(max)
print(range)

################################################################################
splitX=data.sample(frac=0.1,random_state=10)
splitY=data.sample(frac=0.9,random_state=10)

print(splitX)
print(splitY)
print("\n")
################################################################################
print("Mean.median,mode in Scores column")
print("Mean",data["Scores"].mean())
print("Median",data["Scores"].median())
print("Mode",data["Scores"].mode())
print("Standard Deviation",data["Scores"].std())
print("\n")
print("Mean.median,mode in Hours column")
print("Mean",data["Hours"].mean())
print("Median",data["Hours"].median())
print("Mode",data["Hours"].mode())
print("Standard Deviation",data["Hours"].std())

rMean=data.mean(axis=1)
rMedian=data.median(axis=1)
rMode=data.mode(axis=1)
rSD=data.std(axis=1)
print("\n")
print("row_wise mean-\n",rMean)
print("\n")
print("row_wise median-\n",rMedian)
print("\n")
print("row_wise mode-\n",rMode)
print("\n")
print("row_wise std-\n",rSD)
print("\n")

################################################################################
def calVal(num:int):
  return 10*num+0.5

l=data["Hours"].tolist()
 

cal=[]
for i in l:
  a=calVal(i)
  cal.append(a)

exp=data["Scores"].tolist()
print(cal)
print(exp)
mse = np.square(np.subtract(cal,exp)).mean()
print(mse)
```
